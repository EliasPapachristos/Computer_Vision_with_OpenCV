```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
 
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')
 



# When no faces detected, face_classifier returns and empty tuple




#    if cv2.waitKey(15) & 0xFF == 27:
#        break
    
#cv2.destroyAllWindows()
```

```python
img = cv2.imread('Images/eye_face.jpg')
img_fix = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
faces = face_classifier.detectMultiScale(img_fix, 1.3, 5)
```

```python
if faces is ():
    print("No Face Found")
```

```python
def detect_face_eyes(img_fix):
    
    face_rects = face_classifier.detectMultiScale(img_fix)

    for (x,y,w,h) in face_rects:
        cv2.rectangle(img_fix,(x,y),(x+w,y+h),(255,0,0),7)
        #poi_fix = poi_fix[y:y+h, x:x+w]
        #poi_color = img_fix[y:y+h, x:x+w]
        eyes_retcs = eye_classifier.detectMultiScale(img_fix)
        for (ix,iy,iw,ih) in eyes_retcs:
            cv2.rectangle(img_fix,(ix,iy),(ix+iw,iy+ih),(0,0,255),2)
            #cv2.imshow('Face & Eye Detection', img)
            
        #return img_fix
    return img_fix
```

```python
result = detect_face_eyes(img_fix)
plt.imshow(result)
```
